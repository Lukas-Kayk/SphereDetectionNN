```python
# RT-DETR Training + Visualization on Google Colab

# 1. Mount Google Drive
from google.colab import drive
import os
drive.mount('/content/drive')

# 2. Unpack dataset
dataset_zip_path = "/content/drive/MyDrive/dataset.zip"  # Adjust path if needed
!unzip -q "$dataset_zip_path" -d /content/dataset

# 3. Install Ultralytics with RT-DETR
!pip install ultralytics

# 4. Write data.yaml (including test split)
yaml_content = '''
train: /content/dataset/images/train
val: /content/dataset/images/val
test: /content/dataset/images/test

nc: 1
names: ['ball']
'''
with open('/content/data.yaml', 'w') as f:
    f.write(yaml_content)

# 5. RT-DETR Training (comparable to YOLOv5m-level)
!yolo task=detect mode=train model=rtdetr-l.pt data=/content/data.yaml epochs=30 patience=5 imgsz=640 batch=8 name=ball_tracking_rtdetr

# 6. Visualization
from IPython.display import Image, display
import pandas as pd
import matplotlib.pyplot as plt

display(Image(filename='/content/runs/detect/ball_tracking_rtdetr/results.png'))
display(Image(filename='/content/runs/detect/ball_tracking_rtdetr/labels.jpg'))

# 6b. Independent TEST evaluation
print("="*40)
print("TEST EVALUATION (RT-DETR with best.pt)")
print("="*40)

!yolo task=detect mode=val model=/content/runs/detect/ball_tracking_rtdetr/weights/best.pt data=/content/data.yaml split=test

print("="*40)
print("End of test evaluation")
print("="*40)

# 7. Training evaluation – plots from results.csv
print("="*40)
print("TRAINING EVALUATION (from results.csv)")
print("="*40)

results_csv = '/content/runs/detect/ball_tracking_rtdetr/results.csv'
df = pd.read_csv(results_csv)
df.columns = df.columns.str.strip()

# Show available columns
print("Available columns:", df.columns.tolist())

# Dynamic plotting for available metrics (RT-DETR compatible)
plt.figure(figsize=(12, 6))
if 'train/box_loss' in df.columns:
    plt.plot(df['epoch'], df['train/box_loss'], label='Box Loss')
if 'metrics/mAP50(B)' in df.columns:
    plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP@0.5')
if 'metrics/precision(B)' in df.columns:
    plt.plot(df['epoch'], df['metrics/precision(B)'], label='Precision')
if 'metrics/recall(B)' in df.columns:
    plt.plot(df['epoch'], df['metrics/recall(B)'], label='Recall')

plt.xlabel("Epoch")
plt.ylabel("Value")
plt.title("RT-DETR Training Curve")
plt.legend()
plt.grid(True)
plt.show()

print("="*40)
print("End of training evaluation")
print("="*40)

# 8. Save best model to Google Drive
!cp /content/runs/detect/ball_tracking_rtdetr/weights/best.pt /content/drive/MyDrive/best_rtdetr_ball_model.pt

# 9. Check GPU availability
import torch
print("CUDA available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
!nvidia-smi

# 10. Backup the entire runs folder to Google Drive
import shutil
runs_src = '/content/runs/detect'
runs_dst = '/content/drive/MyDrive/rtdetr_runs_backup'
shutil.copytree(runs_src, runs_dst)
print(f"runs folder was copied to {runs_dst}.")

# 11. Example predictions with RT-DETR on test images
from IPython.display import Image, display
from os.path import basename

# Select specific test images (replace with your real ones if needed)
test_imgs = [
    "/content/dataset/images/test/000467.jpg",
    "/content/dataset/images/test/004750.jpg",
    "/content/dataset/images/test/005015.jpg"
]

# Predictions with high confidence threshold
for img_path in test_imgs:
    !yolo task=detect mode=predict model=/content/runs/detect/ball_tracking_rtdetr/weights/best.pt source="{img_path}" save=True conf=0.7 project=pred_output name=test_example exist_ok=True
    display(Image(filename=f"pred_output/test_example/{basename(img_path)}"))
    print(f"Prediction for: {basename(img_path)}")
```
