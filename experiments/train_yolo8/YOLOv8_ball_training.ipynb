```python
# YOLOv8 Training + Visualization on Google Colab

# 1. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Set locale to UTF-8 (to avoid encoding issues in system commands)
import os
import locale
os.environ['LANG'] = 'C.UTF-8'
os.environ['LC_CTYPE'] = 'C.UTF-8'
os.environ['LC_ALL'] = 'C.UTF-8'

try:
    locale.setlocale(locale.LC_ALL, 'C.UTF-8')
except locale.Error:
    print("Warning: Could not set locale to C.UTF-8. System commands might fail.")

# 2. Unpack dataset
zip_path = "/content/drive/MyDrive/dataset.zip"  # Adjust path if needed
!unzip -q "$zip_path" -d /content/dataset

# 3. Install YOLOv8
!pip install ultralytics

# 4. Imports & YAML config
from ultralytics import YOLO
import os
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display

# Write YAML (with train, val, and test splits)
yaml_path = "/content/dataset/data.yaml"
with open(yaml_path, "w") as f:
    f.write("""train: images/train
val: images/val
test: images/test

nc: 1
names: ['ball']
""")

# 5. Training
model = YOLO("yolov8m.pt")
results = model.train(
    data=yaml_path,
    imgsz=640,
    epochs=30,
    batch=8,
    name="ball_tracking_v8",
    project="runs_yolo8",
    patience=5
)

# 6. Visualization
results_path = "/content/runs_yolo8/ball_tracking_v8/results.png"
labels_path = "/content/runs_yolo8/ball_tracking_v8/labels.jpg"
display(Image(filename=results_path))
display(Image(filename=labels_path))

# 6b. Test evaluation
print("="*40)
print("TEST EVALUATION (YOLOv8 with best.pt)")
print("="*40)

!yolo task=detect mode=val model=/content/runs_yolo8/ball_tracking_v8/weights/best.pt data=/content/dataset/data.yaml split=test

print("="*40)
print("End of test evaluation")
print("="*40)

# 7. Training evaluation â€“ plots from results.csv
print("="*40)
print("TRAINING EVALUATION (from results.csv)")
print("="*40)

csv_path = "/content/runs_yolo8/ball_tracking_v8/results.csv"
df = pd.read_csv(csv_path)
df.columns = df.columns.str.strip()

# Print available columns
print("Available columns:", df.columns.tolist())

# Dynamic plotting of available metrics
plt.figure(figsize=(12, 6))
if 'train/box_loss' in df.columns:
    plt.plot(df['epoch'], df['train/box_loss'], label='Box Loss')
if 'train/cls_loss' in df.columns:
    plt.plot(df['epoch'], df['train/cls_loss'], label='Cls Loss')
if 'metrics/mAP50(B)' in df.columns:
    plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP@0.5')
if 'metrics/precision(B)' in df.columns:
    plt.plot(df['epoch'], df['metrics/precision(B)'], label='Precision')
if 'metrics/recall(B)' in df.columns:
    plt.plot(df['epoch'], df['metrics/recall(B)'], label='Recall')

plt.xlabel("Epoch")
plt.ylabel("Value")
plt.title("Training Curve (YOLOv8)")
plt.legend()
plt.grid(True)
plt.show()

print("="*40)
print("End of training evaluation")
print("="*40)

# 8. Save best model to Google Drive
!cp /content/runs_yolo8/ball_tracking_v8/weights/best.pt /content/drive/MyDrive/best_yolov8_m_ball_model.pt

# 9. Check GPU availability
import torch
print("CUDA available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
!nvidia-smi

# 10. Backup the entire runs folder to Google Drive
import shutil

runs_src = '/content/runs_yolo8'
runs_dst = '/content/drive/MyDrive/yolov8_runs_backup'
shutil.copytree(runs_src, runs_dst)

print(f"runs folder was copied to {runs_dst}.")

# 11. Example predictions
from IPython.display import Image
from os.path import basename

# Specific test images (same selection as RT-DETR)
test_imgs = [
    "/content/dataset/images/test/000467.jpg",
    "/content/dataset/images/test/004750.jpg",
    "/content/dataset/images/test/005015.jpg"
]

for img_path in test_imgs:
    !yolo task=detect mode=predict model=/content/runs_yolo8/ball_tracking_v8/weights/best.pt source="{img_path}" save=True conf=0.7 project=pred_output name=test_example exist_ok=True
    display(Image(filename=f"pred_output/test_example/{basename(img_path)}"))
    print(f"Prediction for: {basename(img_path)}")
```
