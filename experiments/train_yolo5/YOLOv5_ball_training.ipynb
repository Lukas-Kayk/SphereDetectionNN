```python
# YOLOv5 Training + Visualization on Google Colab

# 1. Mount Google Drive
from google.colab import drive
import os
drive.mount('/content/drive')

# 2. Unpack dataset
dataset_zip_path = "/content/drive/MyDrive/dataset.zip"  # Adjust if needed
!unzip -q "$dataset_zip_path" -d /content/dataset

# 3. Clone YOLOv5 repository and install dependencies
!git clone https://github.com/ultralytics/yolov5
%cd yolov5
!pip install -r requirements.txt

# 4. Write data.yaml (including test split!)
yaml_content = '''
train: ../dataset/images/train
val: ../dataset/images/val
test: ../dataset/images/test

nc: 1
names: ['ball']
'''
with open('data.yaml', 'w') as f:
    f.write(yaml_content)

# 5. Start training
!python train.py --img 640 --batch 8 --epochs 30 --data data.yaml --weights yolov5m.pt --name ball_tracking_m --patience 5

# 6. Visualize results
from IPython.display import Image, display
import pandas as pd
import matplotlib.pyplot as plt

display(Image(filename='runs/train/ball_tracking_m/results.png'))
display(Image(filename='runs/train/ball_tracking_m/labels.jpg'))

# 6b. Independent TEST evaluation
print("="*40)
print("TEST EVALUATION (independent from training)")
print("="*40)

!python val.py --data data.yaml --weights runs/train/ball_tracking_m/weights/best.pt --task test --img 640

print("="*40)
print("End of test evaluation")
print("="*40)

# 7. Training evaluation ‚Äì plots from results.csv
print("="*40)
print("TRAINING EVALUATION (from results.csv)")
print("="*40)

results_csv = 'runs/train/ball_tracking_m/results.csv'
df = pd.read_csv(results_csv)
df.columns = df.columns.str.strip()

# Show available columns
print("Available columns:", df.columns.tolist())

# Plot only if columns exist
plt.figure(figsize=(12, 6))
if 'train/box_loss' in df.columns:
    plt.plot(df['epoch'], df['train/box_loss'], label='Box Loss')
if 'train/obj_loss' in df.columns:
    plt.plot(df['epoch'], df['train/obj_loss'], label='Obj Loss')
if 'metrics/mAP_0.5' in df.columns:
    plt.plot(df['epoch'], df['metrics/mAP_0.5'], label='mAP@0.5')
if 'metrics/precision' in df.columns:
    plt.plot(df['epoch'], df['metrics/precision'], label='Precision')
if 'metrics/recall' in df.columns:
    plt.plot(df['epoch'], df['metrics/recall'], label='Recall')

plt.xlabel("Epoch")
plt.ylabel("Value")
plt.title("YOLOv5 Training Curve")
plt.legend()
plt.grid(True)
plt.show()

# 8. Save best model to Google Drive
!cp /content/yolov5/runs/train/ball_tracking_m/weights/best.pt /content/drive/MyDrive/best_yolov5_m_ball_model.pt

# 9. Check GPU availability
import torch
print("CUDA available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
!nvidia-smi

# 10. Backup the entire runs folder to Google Drive
import shutil

runs_src = '/content/yolov5/runs'
runs_dst = '/content/drive/MyDrive/yolov5_runs_backup'
shutil.copytree(runs_src, runs_dst)

print(f"runs folder was copied to {runs_dst}.")

# 11. Example predictions
from IPython.display import Image
from os.path import basename

# Specific test images (same as for RT-DETR & YOLOv8)
test_imgs = [
    "/content/dataset/images/test/000467.jpg",
    "/content/dataset/images/test/004750.jpg",
    "/content/dataset/images/test/005015.jpg"
]

for img_path in test_imgs:
    !python detect.py --weights /content/yolov5/runs/train/ball_tracking_m/weights/best.pt --img 640 --conf 0.7 --source {img_path} --project pred_output --name test_example --exist-ok
    display(Image(filename=f"pred_output/test_example/{basename(img_path)}"))
    print(f"Prediction for: {basename(img_path)}")
```
